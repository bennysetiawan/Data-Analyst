```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# grafik scatterplot
plt.scatter(gdp, score)
plt.title("GDP v Happiness Score")
plt.xlabel("GDP/Capita")
plt.ylabel("Happiness Score")
plt.show()
```

```python
# plot histogram chart

plt.hist(score)
plt.title("Happiness core Distribution")
plt.xlabel("Happines Score")
plt.ylabel("Frequenty")
plt.show()
```

```python
# plot bar
# hindari membuat plot dengn data bolean, float, string-tostring(x string, y string)

roundedScore = score.apply(int)

count = roundedScore.value_counts()

hapScore = count.index

count
```

```python
df2 = pd.read_csv("https://dqlab-dataset.s3-ap-southeast-1.amazonaws.com/retail_raw_reduced.csv")
```

```python
df2.head()
```

```python
import datetime
df2['month'] = df2['order_date'].apply(lambda x: datetime.datetime.strptime(x, "%Y-%m-%d").strftime("%Y-%m"))

df2['gmv'] = df2['item_price']* df2['quantity']

df2
```

```python
# multiline chart

df2.groupby(['month', 'brand'])['gmv'].sum().unstack().plot()
plt.title("Breakdown Brand by GMV 2019", loc='center',pad=38, fontsize=28, color='blue')
plt.xlabel("Order Month", fontsize=15)
plt.ylabel("Total Amount (In Billion)", fontsize=15)
plt.grid(color='darkgray', linestyle=':', linewidth=0.5)
plt.ylim(ymin=0)
labels, location = plt.yticks()
plt.yticks(labels, (labels/1000000000).astype(int))
plt.gcf().set_size_inches(12,5)
plt.tight_layout()
plt.legend(loc='right', bbox_to_anchor=(0.5,0.5),shadow=True, ncol=2)
plt.show()
```

```python
plt.clf()

df2.groupby(['month','province'])['gmv'].sum().unstack().plot()
plt.title("Breakdown Brand by GMV 2019", loc='center',pad=30, fontsize=20, color='blue')
plt.xlabel("Order Month", fontsize=15)
plt.ylabel("Total Amount (In Billion)", fontsize=15)
plt.grid(color='darkgray', linestyle=':', linewidth=0.5)
plt.ylim(ymin=0)
labels, location = plt.yticks()
plt.yticks(labels, (labels/1000000000).astype(int))
plt.legend(loc='lower center',bbox_to_anchor=(0.5,--0.5),shadow=True, ncol=3, title='province', fontsize=9, title_fontsize=11)
plt.gcf().set_size_inches(10,5)
plt.tight_layout()
plt.show()
```

```python
top_province = (df2.groupby('province')['gmv']
               .sum()
               .reset_index()
               .sort_values(by='gmv', ascending=False)
               .head(5)
               )
top_province

df2['province_top'] = df2['province'].apply(lambda x: x if(x in top_province['province'].to_list()) else 'other')
df2
```

```python
top_province = (df2.groupby('province')['gmv']
               .sum()
               .reset_index()
               .sort_values(by='gmv', ascending=False)
               .head(5)
               )
top_province

df2['province_top'] = df2['province'].apply(lambda x: x if(x in top_province['province'].to_list()) else 'other')
df2
df2.groupby(['month','province_top'])['gmv'].sum().unstack().plot(marker='.', cmap='plasma')
```

```python
# table temporary -> View Event
# top_provinces <- perbandingan dataset kita dengan table temporary <- global variabel
# Series <- tipe data type -> ini bukan dataset

top_province = (df2.groupby('province')['gmv']
               .sum()
               .reset_index()
               .sort_values(by='gmv', ascending=False)
               .head(5)
               )
top_province

df2['province_top'] = df2['province'].apply(lambda x: x if(x in top_province['province'].to_list()) else 'other')
print(df2)
df2.groupby(['month','province_top'])['gmv'].sum().unstack().plot(marker='.', cmap='plasma')
plt.title('Monthly GMV Year 2019 - Breakdown by Province', loc='center',pad=30, fontsize=20, color='blue')
plt.xlabel("Order Month", fontsize=15)
plt.ylabel("Total Amount (In Billion)", fontsize=15)
plt.grid(color='darkgray', linestyle=':', linewidth=0.5)
plt.ylim(ymin=0)
labels, location = plt.yticks()
plt.yticks(labels, (labels/1000000000).astype(int))
plt.legend(loc='upper center',bbox_to_anchor=(1.1, 1),shadow=True, ncol=1)
plt.gcf().set_size_inches(12,5)
plt.tight_layout()
plt.show()
```

```python
# pie chart
# visualisasi data untuk mempresentasikan hasil perbandingan dari dataset bisa percentage, bisa variabel
# subset data -> exploratory data analysis (EDA)

df_dki_q4 = df2[(df2['province']=='DKI Jakarta')& (df2['month']>='2019-10')]

gmv_per_city_q4 = df_dki_q4.groupby('city')['gmv'].sum().reset_index()
plt.figure(figsize=(6,6))

plt.pie(gmv_per_city_q4['gmv'],labels=gmv_per_city_q4['city'], autopct='%1.2f%%')
plt.title('GMV Contribution Per City - DKI Jakarta in 04 2019', loc='center',pad=30, fontsize=15, color='blue')
plt.show()
```
